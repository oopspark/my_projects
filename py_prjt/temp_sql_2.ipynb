```python
%%sql

DROP VIEW IF EXISTS geodata.country_view;
CREATE VIEW geodata.country_view AS (
    SELECT 
        cc.country,
        rv.iso3,
        rv.region,
        rv.subregion,
        cc.capital,
        cc.latitude,
        cc.longitude
    FROM 
        geodata.country_capital cc
    JOIN
        wpp_2024.regions_view rv 
        ON 
            cc.country = rv.country
);
```

```python
%%sql

select *
from geodata.country_view;
```
