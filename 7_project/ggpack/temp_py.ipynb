```python
from gg import *


# bar 그래프
a = Gg().geom_line().theme_presentation().draw()

# ax = plt.gca()
# a.ax.get_xaxis().set_visible(False)  # x축만 제거
# a.ax.get_yaxis().set_visible(False)  # y축만 제거

for spine in ['top', 'left', 'right']:
    a.ax.spines[spine].set_visible(False)

a.ax.tick_params(left=False, bottom=False)  # y축, x축 틱만 제거


a.ax.grid(axis='y', color='lightgray')  # 가로 그리드만
a.ax.grid(axis='x', visible=False)      # 세로 그리드 제거

a.ax.set_axisbelow(True)

# 축 이름 제거
a.ax.set_xlabel('')  # x축 이름 제거
a.ax.set_ylabel('')  # y축 이름 제거

a.show()
```
